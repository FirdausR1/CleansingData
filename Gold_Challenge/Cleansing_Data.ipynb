```python

def cleansing_data(tweets):
    tweets= tweets.lower()
    tweets=  re.sub(r'&amp', ' ', str(tweets)) #menghapus &amp
    tweets= re.sub(r'user', ' ', str(tweets)) #menghilangkan kata yang double 
    tweets= re.sub(r'#\w+',' ', str(tweets)) #menghapus hastag
    tweets= re.sub(r'\\[0-9A-z]{2,}',' ', str(tweets)) # \\khusus
    tweets= re.sub(r'\\[0-9A-z]',' ', str(tweets)) #menghapus \\n
    tweets= re.sub(r'[^a-zA-Z\s]',' ', str(tweets)) #menghilangkan huruf dengan karakter khusus
    tweets= re.sub(r'[^\w\s]',' ', str(tweets)) #menghilangkan tanda baca
    tweets= re.sub(r'\s+', ' ', str(tweets)) #menghapus spasi
    tweets= re.sub(r'\t+', '\t ', str(tweets)) #menghapus tab
    tweets= re.sub(r'RT[\s]+', ' ', str(tweets)) #menghapus RT
    tweets= re.sub(r'rt[\s]+', ' ', str(tweets)) #menghapus rt
    tweets= tweets.strip()
    
    
    return tweets

```

```python
kamusalay_new=new_kamusalay.rename(columns={'anakjakartaasikasik': 'Original',
                                 'anak jakarta asyik asyik': 'replace'
                             })
```

```python
alay_dict = dict(zip(kamusalay_new['Original'], kamusalay_new['replace']))
```

```python
kamusalay_new.head()
```

```python
def remove_kata_alay(kata):
    return ' '.join([alay_dict[word] if word in alay_dict else word for word in kata.split(' ')])
```

```python
for kata in df['Tweet']:
    print(remove_kata_alay(kata))
```

```python
kamus_alay_new = {}
for k,v in kamusalay_new.values:
    kamus_alay_new[k] = v

list_abusive = abusive['ABUSIVE'].tolist()

```

```python
list_abusive
```

```python
def processing_word(input_text):
    new_text = [] # set up new list
    new_new_text = [] # set up new new list
    input_text = cleansing_data(input_text)
    text = input_text.split(" ") # split input_text menjadi list of words
    for word in text: # untuk setiap word in 'text'
        if word in list_abusive: # check word di dalam list_of_abusive_words
            continue # jika ada, skip
        else:
            new_text.append(word) # jika tidak ada, masukkan ke dalam list new_text
   
    for word in new_text:
        new_word = kamus_alay_new.get(word, word) # check ke new_kamus_alay, apakah word ada di dictionarynya. kalau ga ada, return word yang sama. kalau ada, kembalikan value barunya (value yang ada di dict)
        new_new_text.append(new_word)
    
    text = " ".join(new_new_text)
    
    return text
```

```python
for data_baru in df['Tweet']:
    print(processing_word(data_baru))
```

```python
df['clean_text'] = df['Tweet'].apply(lambda x: processing_word(x))
df[['Tweet','clean_text']]
```
